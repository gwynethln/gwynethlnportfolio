# Retrieving used notes by brand, comparing use frequency of each <a id='viz'></a>

For simplicity, I won't list all notes considered by all brands. I will use Di Ser's perfumes.

```python
di_ser
```

```python
ds_notes = di_ser['Notes'].str.split()
each_ds_note = set() # using 'set' to UNIQUELY list each note that occurs

for note in ds_notes:
    each_ds_note.update(note) # updating the Set with each note occurring in Di Ser's Notes column
    
list_ds_notes = list(each_ds_note)
list_ds_notes
```

```python
lower_all = [n.lower() for n in list_ds_notes] 
remove_comma = [c.replace(',','') for c in lower_all]
remove_ptheses = [p for p in remove_comma if '(' not in p and ')' not in p]

raw_list = sorted(remove_ptheses) 
print('List of Di Ser notes with duplicates')
raw_list
```

I'll remove adjectives/modifiers of notes to shorten the list. For example, "rose absolute" and "rose otto" would fall under the "rose" category. Same for "jasmine sambac" and "jasmine grandiflorum" for the "jasmine" category.

I'll make more sophisticated code for this purpose later.

```python
to_remove = ['absolute','blossom','sweet','spicy','blue','pink','kyara','grade','green','grandiflorum','otto','sambac','japanese','seed']

for r in to_remove:
    if r in raw_list:
        raw_list.remove(r)
```

```python
raw_list
```

I see more modifiers that should be removed (due to the structure of my for-loop earlier). I'll run this again.

```python
for r in to_remove:
    if r in raw_list:
        raw_list.remove(r)
```

```python
raw_list
```

The above list looks better.

Now, I'll make a clean list of notes without duplicates to prepare for the next task.

```python
clean_list = list(sorted(set(raw_list)))
```

```python
clean_list
```

## Graphing how often each note is used by brand

```python
# counting the frequency of each word used
note_freq = []

for n in clean_list:
    note_freq.append(raw_list.count(n))
print(f'frequency of each note: {note_freq}')
```

### Bad example: BAR CHART

```python
print(f'Number of different notes: {len(clean_list)}')
```

There are 44 different notes included by Di Ser's perfumes. A bar chart (horizontal or vertical) would not provide efficient readability for the viewer.

I'll create a bar chart below, both to show this difficulty and for practice.

```python
# Preparing for Matplotlib Bar Chart
x_notes = np.array(clean_list)
y_freq = np.array(note_freq)

plt.barh(x_notes, y_freq)

plt.yticks(fontsize = 5)

plt.title('Notes used by Di Ser perfumes')
plt.xlabel('Number of times used')
plt.ylabel('Note')

# forcing x-axis 'Number of times used' to show only integer values
x_ax = plt.gca()
x_ax.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.tight_layout()
plt.show()
```

Even if we spaced out the y-axis of notes, it would be inefficient to locate how many times each note was used.

### Better example: Pivot table

What if a pivot table would be better, where we can immediately see the frequency of each note's use immediately? Still, with 44 different notes, it may take a long time to read through the table. I'll try it out first.

```python
perfume_data = {'Di Ser Note' : clean_list, 'Number of times used' : note_freq}
perf_df = pd.DataFrame(perfume_data)
perf_pivot = perf_df.pivot_table(index = 'Di Ser Note', values = 'Number of times used')
perf_pivot
```

This is a pretty long list to scroll through, but this numerical visualization is better than its bar chart. Let's try to graph this, though.

### Better Example 2: Clean Bar Chart - Isolating notes used multiple times vs. those used once

```python
mult_use = ['all other notes (see list below)'] # prepared arrays: showed that all other notes had frequency 1
mult_use_count = [1]

used_once = [] # preparing once-used count of notes for reader reference

for n in clean_list:
    if raw_list.count(n) > 1:
        mult_use.append(n)
        mult_use_count.append(raw_list.count(n))
    else: 
        used_once.append(n)
        
print(f'Notes used (refined): {mult_use}')
print('')
print(f'Number of times used: {mult_use_count}')
```

```python
x_mult = np.array(mult_use[::-1]) # reversed mult_use array for top-to-bottom alphabetical reading order
y_mult_freq = np.array(mult_use_count[::-1]) # same with mult_use_count to align values with correct notes

plt.barh(x_mult, y_mult_freq, color = 'thistle', height = 0.6)
x_ax = plt.gca()
x_ax.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.title('Notes used by Di Ser perfumes')
plt.xlabel('Number of times used')
plt.ylabel('Note')

plt.show()
```

These are the notes used <b>only once</b> by Di Ser, according to the perfume dataset provided.

```python
print('All other notes:')
used_once
```

To compare, I'll show the former and final plots side-by-side.

```python
fig, axes = plt.subplots(1,2, figsize = (18,9), gridspec_kw = {'hspace':0, 'wspace':.9})

# OLD PLOT
axes[0].barh(x_notes, y_freq)
axes[0].xaxis.set_major_locator(MaxNLocator(integer=True))

axes[0].set_title('Old plot', fontsize=17)
axes[0].tick_params(axis='both', labelsize=13)
axes[0].set_xlabel('Number of times used', fontsize=15, labelpad = 15)
axes[0].set_ylabel('Note', fontsize=15)

# NEW PLOT
axes[1].barh(x_mult, y_mult_freq, color = 'thistle', height = 0.6)
axes[1].xaxis.set_major_locator(MaxNLocator(integer=True))

axes[1].set_title('New plot', fontsize=17)
axes[1].tick_params(axis='both', labelsize=13) # adjusting axes tick label font size
axes[1].set_xlabel('Number of times used', fontsize=15, labelpad = 15) # moving xlabel further away from x-axis
axes[1].set_ylabel('Note', fontsize=15)

# Entire grid
plt.suptitle('Comparison: Bar charts of notes used by Di Ser perfumes', fontsize=20)
plt.show()
```
